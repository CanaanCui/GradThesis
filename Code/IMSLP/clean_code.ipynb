```python
import pandas as pd
import re
from datetime import datetime

# 定义一个函数用于从 Life Span 中提取出生和去世日期
def extract_dates(life_span):
    # 使用正则表达式提取日期部分
    dates = re.findall(r'(\d{4})年(\d{1,2})月(\d{1,2})日', life_span)
    
    # 将提取的日期转换为日期格式
    birth_date = datetime(year=int(dates[0][0]), month=int(dates[0][1]), day=int(dates[0][2]))
    death_date = datetime(year=int(dates[1][0]), month=int(dates[1][1]), day=int(dates[1][2]))
    
    return birth_date.strftime('%Y-%m-%d'), death_date.strftime('%Y-%m-%d')

# 使用 apply 方法将 Life Span 中的出生和去世日期提取出来并分成两列
df['Birth Date'], df['Death Date'] = zip(*df['Life Span'].apply(extract_dates))

df.head()
```
